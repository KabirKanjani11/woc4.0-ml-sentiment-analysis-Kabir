```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import pandas as pd
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
df=pd.read_csv('tripadvisor_hotel_reviews.csv')
ps=PorterStemmer()
def func(x):

    if x>=0 and x<=2:
        return 0
    elif x == 3:
        return 1
    elif x >3 and x<=5:
        return 2
    else:
        print(x)
        return -1
```

```python
df.iloc[:,1]=df.iloc[:,1].apply(lambda  x:int(x))
df.iloc[:,1]=df.iloc[:,1].apply(func)
df
```

```python
#Convert String to lower
df.iloc[:,0]=df.iloc[:,0].str.lower()
df
```

```python
#Remove all the hyperlinks if present in the text.
df.iloc[:,0]=df.iloc[:,0].apply(lambda x:re.sub(r'[(https|http)://]?[www\.]?[a-zA-Z0-9]+\.[(com)|(org)|(edu)]+','',str(x)) for x in df['Review'])
df
```

```python
#Substitute all the non alphabetical characters(a-z) by ‘ ‘ (space).
df.iloc[:,0]=df.iloc[:,0].apply(lambda x:re.sub(r'[(https|http)://]?[www\.]?[a-zA-Z0-9]+\.[(com)|(org)|(edu)]+','',str(x)) for x in df['Review'])
df
```

```python
#If there are multiple consequent spaces in the text, replace them by a single space (for example ‘    ‘ to ‘ ‘ ).
df.iloc[:,0]=df.iloc[:,0].apply(lambda x:re.sub(re.sub(r'[^a-z & ^\s]','',str(x)) for x in df['Review'])
df
```

```python
df.iloc[:,0]= df.iloc[:,0].apply(lambda x: word_tokenize(x))
stop_words = set(stopwords.words('english'))
#it will take time to load 
```

```python
def remover(p):
    for rword in p:
        if rword in stop_words:
            p.remove(rword)
    return p


df['Review'] = df['Review'].map(remover)
```
