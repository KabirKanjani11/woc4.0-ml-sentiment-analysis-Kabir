```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
lc=LabelEncoder()
```

```python
data=pd.read_csv('melb_data.csv')
```

```python
data.iloc[:,0]=data.iloc[:,0].apply(lambda x:x.replace("ID_",""))
```

```python
data.iloc[:,0]=data.iloc[:,0].apply(lambda x:int(x))
```

```python
data.set_index(data.columns[0])
```

```python
data['Suburb']=lc.fit_transform(data['Suburb'])
data['Type']=lc.fit_transform(data['Type'])
data['SellerG']=lc.fit_transform(data['SellerG'])
data['CouncilArea']=lc.fit_transform(data['CouncilArea'])
data['Regionname']=lc.fit_transform(data['Regionname'])
data
```

```python
data.interpolate()
```

```python
data=data.fillna(0)
data.iloc[:,10].apply(lambda x:int(x))
data.iloc[:,13].apply(lambda x:int(x))
data.iloc[:,15].apply(lambda x:int(x))
data.iloc[:,16].apply(lambda x:int(x))
data
```

```python
print("Mean data for rooms",data['Rooms'].mean())
print("Median data for rooms",data['Rooms'].median())
print("Mode data for Rooms",data['Rooms'].mode())
print("Mean data for Price",data['Price'].mean())
print("Median data for price",data['Price'].median())
print("Mode data for price",data['Price'].mode())
```
